**#  Sales behavior before, during, and after holidays**

```python
#  Sales behavior before, during, and after holidays
train_df["IsHoliday"] = train_df["StateHoliday"].apply(lambda x: 1 if x in ["a", "b", "c"] else 0)
holiday_sales = train_df.groupby("IsHoliday")["Sales"].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=holiday_sales.index, y=holiday_sales.values)
plt.xticks(ticks=[0, 1], labels=["Non-Holiday", "Holiday"])
plt.ylabel("Average Sales")
plt.title("Sales Behavior on Holidays vs. Non-Holidays")
plt.show()
```

```python
train_df["Date"] = pd.to_datetime(train_df["Date"]) 
train_df["Month"] = train_df["Date"].dt.month  
```

```python
monthly_sales = train_df.groupby("Month")["Sales"].mean()

plt.figure(figsize=(8, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker="o")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.title("Seasonal Trends in Sales")
plt.xticks(range(1, 13))
plt.show()
```

```python
# Correlation between sales and number of customers


corr_sales_customers = train_df[["Sales", "Customers"]].corr().iloc[0, 1]

plt.figure(figsize=(6, 4))
sns.scatterplot(x=train_df["Customers"], y=train_df["Sales"], alpha=0.5)
plt.xlabel("Customers")
plt.ylabel("Sales")
plt.title(f"Correlation Between Sales and Customers: {corr_sales_customers:.2f}")
plt.show()
```

```python
# Effect of promotions on sales

promo_sales = train_df.groupby("Promo")["Sales"].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=promo_sales.index, y=promo_sales.values)
plt.xticks(ticks=[0, 1], labels=["No Promo", "Promo"])
plt.ylabel("Average Sales")
plt.title("Impact of Promotions on Sales")
plt.show()
```

```python
# Customer behavior when stores are open/closed


open_sales = train_df.groupby("Open")["Sales"].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=open_sales.index, y=open_sales.values)
plt.xticks(ticks=[0, 1], labels=["Closed", "Open"])
plt.ylabel("Average Sales")
plt.title("Sales When Store is Open vs. Closed")
plt.show()
```

```python
# Assortment type impact on sales
assortment_sales = train_df.merge(store_df, on="Store").groupby("Assortment")["Sales"].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=assortment_sales.index, y=assortment_sales.values, order=["a", "b", "c"])
plt.xlabel("Assortment Type")
plt.ylabel("Average Sales")
plt.title("Effect of Assortment Type on Sales")
plt.show()
```

```python
#  Distance to competitor vs. sales
train_store_df = train_df.merge(store_df, on="Store")
plt.figure(figsize=(6, 4))
sns.scatterplot(x=train_store_df["CompetitionDistance"], y=train_store_df["Sales"], alpha=0.5)
plt.xlabel("Competition Distance")
plt.ylabel("Sales")
plt.title("Effect of Competitor Distance on Sales")
plt.xlim(0, train_store_df["CompetitionDistance"].quantile(0.99))  # Remove extreme outliers
plt.show()
```

```python
#  New competitors opening vs. sales impact
train_store_df["HasNewCompetitor"] = train_store_df["CompetitionOpenSinceYear"] > 2013  # Assume new competitors after 2013
new_competitor_sales = train_store_df.groupby("HasNewCompetitor")["Sales"].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=new_competitor_sales.index, y=new_competitor_sales.values)
plt.xticks(ticks=[0, 1], labels=["No New Competitor", "New Competitor"])
plt.ylabel("Average Sales")
plt.title("Impact of New Competitors on Sales")
plt.show()
```
